```python
# 1. Widgets for Parameters
dbutils.widgets.text("run_date", "2026-01-15")
v_date = dbutils.widgets.get("run_date")

from pyspark.sql.functions import current_timestamp, lit

file_path = "/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv"
# 2. Read Raw Data (Adjust path to your source)
raw_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)

# 3. Add Metadata (Ingestion Time and Batch Date)
bronze_df = raw_df.withColumn("ingestion_timestamp", current_timestamp()) \
                  .withColumn("batch_date", lit(v_date))

# 4. Write to Bronze Table
bronze_df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ecommerce_bronze")

print(f"Bronze layer complete for {v_date}")
```
